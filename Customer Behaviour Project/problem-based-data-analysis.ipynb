By checking, I have found that **'Review Rating'** has a total of **37** null values. To go to further analysis, I have to do something about the null values.

## Approach to Mitigate the null values.
I can do mean or median for the null values. But there is a problem. For Mean, it can easily be effected by the outliers. For Median, it is robust to outliers. I can replace the null values with overall "Median" of the entire column. In logic, It will be okay but there is a catch. All the null values will be replaced by a generalized median, which can introduce bias to the dataset. But in dataset, we have different types of catagories. So, rather than replacing with general median, I can Replace it by catagory wise median. That way I can get the genuine results.

````python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
````

````python
# checking after the action
df.isnull().sum()
````

# Transforming Dataset for Data Analysis
In the dataset, all the column names has spaces and mixture of upper case  and  lower case characters. This will create a problem during the coding session. So, to fix this I have transformed the column names into snake case.

For Example: We have a column called "Customer ID" insted of writing this we can write "customer_id" to reduce the human error and easier typing.

````python
# checking all the column names
df.columns
````

To change all the column names to the "Snake Case" I first transform all the column names into lower case and the replaced all the whiitespace to '_'.

**Note: Lastly I have replaced 'purchase_amount_(usd)' to 'purchase_amount'**

````python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns= {'purchase_amount_(usd)' : 'purchase_amount'})
````

````python
# after changing
df.columns
````

## Feature Engineering
I have created new columns using feature Engineering called "age_group" which will consist of customers in groups of "Young Adult","Adult","Middle-aged" and "Senior".

".qcut()" method is used for the grouping.

* **Why this new column?** -> To understand easily about the customer age group and for marketing stragies it is helpfull.

````python
# create a column 'age_group'
labels = ["Young Adult","Adult","Middle-aged", "Senior"]
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
````

````python
df[['age', 'age_group']].head(10)
````

I have created another column called "purchase_frequency_days".
* **Why create this column?** -> because we have "frequency_of_purchases" which tells about the purchase, but in text form. That is why this new column.

````python
frequency_mapping = {
    "Fortnightly" : 14,
    "Weekly" : 7,
    "Monthly" : 30,
    "Quarterly" : 90,
    "Bi-Weekly" : 14,
    "Annually" : 365,
    "Every 3 Months" : 90
}

# creating column and the records under the column.
df["purchase_frequency_days"] = df["frequency_of_purchases"].map(frequency_mapping)
````

### Pushing Data to My Local Database

I ran into a problem trying to connect this Kaggle notebook directly to my local PostgreSQL database. Because Kaggle runs on a cloud server, using `localhost` in the connection string just threw a "Connection refused" error. Kaggle couldn't see my laptop's private network.

To get around this, I used a free tunneling tool called **Bore**. It creates a secure, temporary bridge between my local computer and the internet.

**Here is how I set it up:**

1. I ran this command in my computer's terminal to expose my local Postgres port (5432):
```bash
bore local 5432 --to bore.pub

```


2. Bore gave me a public address (`bore.pub`) and a random 5-digit port number.
3. I updated the Python connection string to use that new host and port instead of `localhost`.

**Fixing the Timeout Issue**
Once the connection was working, I hit a `PendingRollbackError` when trying to upload the dataframe. This happened because the public tunnel is slower than a normal local network. Trying to shove the entire dataset through at once caused the connection to choke and time out.

The fix was simple: I added `chunksize=500` to the `to_sql()` function. This tells pandas to send the data over in small batches of 500 rows, which kept the tunnel stable and successfully loaded everything into my database.

**The Final Code:**

```python
from sqlalchemy import create_engine

# Using the temporary credentials from the Bore tunnel
username = 'postgres' 
password = '12345' 
host = 'bore.pub' 
port = '19728' # The port Bore assigned me
database = 'customer_behavior' 

engine = create_engine(f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}")
table_name = "customer"

try:
    # chunksize=500 is required to prevent the tunnel from timing out
    df.to_sql(table_name, engine, if_exists="replace", index=False, chunksize=500)
    print("Data successfully loaded!")
except Exception as e:
    print(f"An error occurred: {e}")
finally:
    engine.dispose() # Close the connection cleanly

```
